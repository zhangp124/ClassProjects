```python
def analyze_entities_with_gemini(text):

    #Prompt given to Gemini
    prompt = f"""
    Analyze the following text. Extract all named entities and categorize them (e.g., person, organization, location, date). Then briefly explain their role or importance in the context of the sentence:

    {text}
    """

    #Model used in this experiement
    model = genai.GenerativeModel("gemini-1.5-flash")
    response = model.generate_content(prompt)
    print(response.text)
```

```python
def analyze_entities(text):

    #Initialize the LanguageService Client
    client = language_v1.LanguageServiceClient()

    #Create a Document object to send to API
    document = language_v1.Document(content=text, type_=language_v1.Document.Type.PLAIN_TEXT)

    #Analyze said entities within "document"
    response = client.analyze_entities(document=document)

    #Print out the recognized entities and their types
    for entity in response.entities:
        print(f"Entity: {entity.name}, Type: {entity.type_}")




```

```python
#Other sample text I decided to test the AI on iterating between the three texts chosen
texts = [
    "Baruch College was established by the City University of New York in 1919 in New York City.",
    "Baruch College was started back in 1919 by CUNY and named after Bernard Baruch, a well-known businessman and political advisor — now it’s a popular school in NYC where people from all over come to study business, public affairs, and the arts.",
    "Baruch College, established in 1919 by the City University of New York and named after financier and presidential advisor Bernard Baruch, has grown into a prominent institution known for its schools of business, public affairs, and liberal arts, attracting students and faculty from around the world to its campus in New York City."
]

#Analyze example text with entities
sample_text = "Baruch College was established by the City University of New York in 1919 in New York City."

for i, text in enumerate(texts, 1):
    print(f"\n--- Analyzing text{i} ---")
    analyze_entities_with_gemini(text)

```
